```python
from Class_PINN_FBPINN import *
import matplotlib.pyplot as plt

# Common parameters
domain_extrema = [-2*torch.pi, 2*torch.pi]
activation_function = nn.Tanh()

FIGSIZE = (6, 4)

SAVE_FIGURES = False
```

# Figure 1

## PINN --> w = 1, n_hidden_layers = 2, neurons = 16

```python
w = 1
n_hidden_layers = 2
neurons = 16

num_points = 200
n_epochs = 50000

# Create the model
model = PINN_Cos_1D(domain_extrema, n_hidden_layers, neurons, activation_function, w)

# Load the parameters
model.load_state_dict(torch.load(f'Models/PINN_w_{w}_n_hidden_layers_{n_hidden_layers}_neurons_{neurons}.pt'))
history = np.loadtxt(f'Models/PINN_w_{w}_n_hidden_layers_{n_hidden_layers}_neurons_{neurons}_history.txt')
l1_loss_1 = np.loadtxt(f'Models/PINN_w_{w}_n_hidden_layers_{n_hidden_layers}_neurons_{neurons}_l1_loss.txt')
```

```python
# Plot
model.eval()

# test points
x = torch.linspace(domain_extrema[0], domain_extrema[1], 1000, dtype=torch.float32, device=DEVICE).reshape(-1, 1)

# Evaluate the exact solution
u_exact = model.exact_solution(x)

# Evaluate the approximate solution
u_approx = torch.tanh(model.w * model.normalize_input(x)) * model.unnormalize_output( model( model.normalize_input(x)) )

# Resale the approximate solution to the same scale as the exact solution
u_approx = model.restore_output(u_approx)

#Plot
plt.figure(figsize=FIGSIZE)
plt.plot(x.cpu().detach().numpy(), u_exact.cpu().detach().numpy(), label='Exact Solution')
plt.plot(x.cpu().detach().numpy(), u_approx.cpu().detach().numpy(), label='PINN')
plt.title(f'(a) PINN (w={model.w}, {model.n_hidden_layers} layers, {model.neurons[0]} hidden units)')
plt.xlabel(r'$x$')
plt.ylabel(r'$u$')
plt.legend()
if SAVE_FIGURES: plt.savefig('Plots/figure_1_a.pdf', bbox_inches='tight')

```

# Figure 5

## FBPINN --> w = 1, n_hidden_layers = 2, neurons = 16 n_subdomains = 5, overlap = 1.3, sigma = 0.1

```python
w = 1
n_hidden_layers = 2
neurons = 16

n_subdomains = 5
overlap = 1.3
sigma = 0.1

num_points = 200
n_epochs = 50000

# Create the model
model = FBPPINN_Cos_1D(domain_extrema=domain_extrema, n_subdomains=n_subdomains, overlap=overlap, sigma=sigma, n_hidden_layers=n_hidden_layers, neurons=neurons, activation_function=nn.Tanh(), w=w)

# Load the parameters
for i in range(n_subdomains):
    model.neural_networks[i].load_state_dict(torch.load(f'Models/FBPINN_w_{w}_n_hidden_layers_{n_hidden_layers}_neurons_{neurons}_n_subdomains_{n_subdomains}_overlap_{overlap}_sigma_{sigma}_subnet_{i}.pt'))
history = np.loadtxt(f'Models/FBPINN_w_{w}_n_hidden_layers_{n_hidden_layers}_neurons_{neurons}_n_subdomains_{n_subdomains}_overlap_{overlap}_sigma_{sigma}_history.txt')
l1_loss_5 = np.loadtxt(f'Models/FBPINN_w_{w}_n_hidden_layers_{n_hidden_layers}_neurons_{neurons}_n_subdomains_{n_subdomains}_overlap_{overlap}_sigma_{sigma}_l1_loss.txt')
```

```python
fig = plt.figure(figsize=FIGSIZE)

x = torch.linspace(domain_extrema[0], domain_extrema[1], num_points, dtype=torch.float32, device=DEVICE).reshape(-1, 1)

for i in range(model.n_subdomains):
    NN_i = model.neural_networks[i]
    NN_i.eval()

    x_test_norm = model.normalize_input(x)
    window_function = model.window_function(x, model.midpoints_overlap[i][0], model.midpoints_overlap[i][1])
    u_i =  window_function * model.unnormalize_output( NN_i( x_test_norm ) )
    
    u_approx_i = torch.tanh(model.w * x) * u_i


    plt.plot(x.cpu().detach().numpy(), u_approx_i.cpu().detach().numpy())
plt.title('(a) FBPINN (individual network solutions)')
plt.xlabel(r'$x$')
plt.ylabel(r'$u$')
plt.legend()
if SAVE_FIGURES: plt.savefig('Plots/figure_5_a.pdf', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=FIGSIZE)

x = torch.linspace(domain_extrema[0], domain_extrema[1], num_points, dtype=torch.float32, device=DEVICE).reshape(-1, 1)

u_exact = model.exact_solution(x)
u_approx = torch.tanh(model.w * x) * model(x)
plt.plot(x.cpu().detach().numpy(), u_exact.cpu().detach().numpy(), label='Exact', color="blue")
plt.plot(x.cpu().detach().numpy(), u_approx.cpu().detach().numpy(), label='FBPINN', color="green")


plt.title('(b) FBPINN (full solution)')
plt.xlabel(r'$x$')
plt.ylabel(r'$u$')
plt.legend()
if SAVE_FIGURES: plt.savefig('Plots/figure_5_b.pdf', bbox_inches='tight')
```

```python
# PLOT THE WINDOW FUNCTIONS AND THE SUBDOMAINS
subdomains = model.subdomains
midpoints = model.midpoints_overlap

# Create a figure with three subplots in the same frame
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, gridspec_kw={'hspace': 0}, figsize=FIGSIZE)

# The window functions
window_fct = []
x = np.linspace(domain_extrema[0], domain_extrema[1], 1000)
for i in range(model.n_subdomains):
    window_fct.append(model.window_function(x, midpoints[i][0], midpoints[i][1]))

# fig = plt.figure(figsize=(10, 5))
for i in range(model.n_subdomains):
    ax1.plot(x, window_fct[i], linewidth=2)

# The subdomain
for i in range(model.n_subdomains):
    x = np.linspace(subdomains[i][0], subdomains[i][1], 100)
    y = np.full(100, i%2*0.01)
    ax2.plot(x, y, linewidth=4)
ax2.set_ylim([-0.1, 0.1])

# The overlap
x = np.linspace(domain_extrema[0], domain_extrema[1], 1000)
ax3.plot(x, np.full(1000, 0.5), linewidth=4, color="gray")
for i in range(1, model.n_subdomains):
    overlap_i = np.linspace(subdomains[i][0], subdomains[i-1][1], 100)
    ax3.plot(overlap_i, np.full(100, 0.5), linewidth=6, color="hotpink")




# Remove y-axis tick labels and ticks on the second and third subplot
ax1.set_yticks([0,0.5,1])
ax1.set_yticklabels([0, "Window\nfunction" ,1])

ax2.set_yticks([0])
ax2.set_yticklabels(["Subdomain\ndefinition"])

ax3.set_yticks([0.5])
ax3.set_yticklabels(["Overlap\nmodels"])

fig.suptitle('(c) FBPINN subdomain definition and window functions')
plt.xlabel(r'$x$')
if SAVE_FIGURES: plt.savefig('Plots/figure_5_c.pdf', bbox_inches='tight')
```
